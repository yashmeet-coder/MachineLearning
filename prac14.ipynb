```python
'''
Consider the dataset iris.
Apply hierarchical clustering algorithm to cluster the dataset into 3 clusters.
'''
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
X = iris.data
```

```python
clustering = AgglomerativeClustering(n_clusters=3)
cluster_labels = clustering.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=cluster_labels, palette=sns.color_palette('colorblind', 1))
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hierarchical Clustering of Iris Dataset')
plt.legend(title='Cluster')
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=2)
cluster_labels = clustering.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=cluster_labels, palette=sns.color_palette('colorblind', 2))
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hierarchical Clustering of Iris Dataset')
plt.legend(title='Cluster')
plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=3)
cluster_labels = clustering.fit_predict(X)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=cluster_labels, palette=sns.color_palette('colorblind', 3))
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hierarchical Clustering of Iris Dataset')
plt.legend(title='Cluster')
plt.show()
```
