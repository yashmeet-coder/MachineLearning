```python
'''
Consider the dataset california_housing from sklearn. 
Find the correlation b/w the different attributes of this dataset.
Using the least square estimation method from sklearn, find the best fit line. 
Also find the error.
'''
```

```python
from sklearn.datasets import fetch_california_housing
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
california_housing = fetch_california_housing()
X, y = california_housing.data, california_housing.target
```

```python
correlation_matrix = np.corrcoef(X, rowvar=False)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", 
            xticklabels=california_housing.feature_names, yticklabels=california_housing.feature_names)
plt.title('Correlation Heatmap of California Housing Attributes')
plt.show()

regressor = LinearRegression()
regressor.fit(X, y)

y_pred = regressor.predict(X)

mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```
