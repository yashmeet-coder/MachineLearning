```python
'''
Consider the dataset ‘Adveristing.csv’. Find the correlation coefficient between the input
attributes TV, Radio, Newspaper and Output Attribute Sales. Use least square estimation
method to find the line of regression b/w
1. TV and Sales
2. Radio and Sales
3. Newspaper and Sales
For all of the above options, also draw a scatter plot and line of regression.
Also find the error in each of the above.
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('Advertising.csv')

correlation_tv_sales = data['TV'].corr(data['Sales'])
correlation_radio_sales = data['Radio'].corr(data['Sales'])
correlation_newspaper_sales = data['Newspaper'].corr(data['Sales'])

print("Correlation between TV and Sales:", correlation_tv_sales)
print("Correlation between Radio and Sales:", correlation_radio_sales)
print("Correlation between Newspaper and Sales:", correlation_newspaper_sales)

X_tv = data['TV'].values.reshape(-1, 1)
X_radio = data['Radio'].values.reshape(-1, 1)
X_newspaper = data['Newspaper'].values.reshape(-1, 1)
y = data['Sales'].values
```

```python
regressor_tv = LinearRegression()
regressor_tv.fit(X_tv, y)
y_pred_tv = regressor_tv.predict(X_tv)
mse_tv = mean_squared_error(y, y_pred_tv)
print("Mean Squared Error (TV vs Sales):", mse_tv)
```

```python
regressor_radio = LinearRegression()
regressor_radio.fit(X_radio, y)
y_pred_radio = regressor_radio.predict(X_radio)
mse_radio = mean_squared_error(y, y_pred_radio)
print("Mean Squared Error (Radio vs Sales):", mse_radio)
```

```python
regressor_newspaper = LinearRegression()
regressor_newspaper.fit(X_newspaper, y)
y_pred_newspaper = regressor_newspaper.predict(X_newspaper)
mse_newspaper = mean_squared_error(y, y_pred_newspaper)
print("Mean Squared Error (Newspaper vs Sales):", mse_newspaper)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_tv, y, color='blue', label='Data points')
plt.plot(X_tv, y_pred_tv, color='red', linewidth=2, label='Regression line')
plt.xlabel('TV Advertising ($)')
plt.ylabel('Sales')
plt.title('TV Advertising vs Sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_radio, y, color='blue', label='Data points')
plt.plot(X_radio, y_pred_radio, color='red', linewidth=2, label='Regression line')
plt.xlabel('Radio Advertising ($)')
plt.ylabel('Sales')
plt.title('Radio Advertising vs Sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_newspaper, y, color='blue', label='Data points')
plt.plot(X_newspaper, y_pred_newspaper, color='red', linewidth=2, label='Regression line')
plt.xlabel('Newspaper Advertising ($)')
plt.ylabel('Sales')
plt.title('Newspaper Advertising vs Sales')
plt.legend()
plt.show()
```
